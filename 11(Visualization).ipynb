## 2. 간단한 선 그래프 그리기

```python
# $ mkdir -p ~/aiffel/data_visualization
# $ ln -s ~/data ~/aiffel/data_visualization/
```

```python
from datetime import datetime
import pandas as pd
import os

# 그래프 데이터
csv_path = os.getenv('HOME') + "/aiffel/data_visualization/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']

# 축 그리기 및 좌표축 설정
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
price.plot(ax=ax, style='black')
plt.ylim([1600, 2200])
plt.xlim(['2019-05-01', '2020-03-01'])

# 주석 달기
important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020, 2, 19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d) + 10), # 주석을 달 좌표 (x, y)
               xytext = (d, price.asof(d) + 100), # 주석 텍스트가 위치할 좌표 (x, y)
               arrowprops = dict(facecolor='red')) # 화살표 추가 및 색 설정

# 그리드, 타이틀 달기
plt.grid()
ax.set_title('StockPrice')

# 보여주기
plt.show()
```

## 3. plot 사용법 상세
- 위에서는 figure() 객체 생성하고, add_subplot()으로 서브플롯을 생성해 plot을 그렸다.
    - -> plt.plot() 명령으로 그래프를 그리면 생략 가능

##### plt.plot()로 그래프 그리기

```python
import numpy as np
x = np.linspace(0, 10, 100) # 0에서 10까지 균등한 간격으로 100개의 숫자 만들라
plt.plot(x, np.sin(x), 'o')
plt.plot(x, np.cos(x), '--', color='black')
plt.show()
```

##### 서브 플롯

```python
x = np.linspace(0, 10, 100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), 'orange', 'o')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'orange')
plt.show()
```

##### linestyle, marker 옵션

```python
x = np.linspace(0, 10, 100)

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+0, '-g') # solid green
plt.plot(x, x+1, '--c') # dashed cyan
plt.plot(x, x+2, '-.k') # dashdot black
plt.plot(x, x+3, ':r') # dotted red

plt.plot(x, x+4, linestyle='-') # solid
plt.plot(x, x+5, linestyle='--') # dashed
plt.plot(x, x+6, linestyle='-.') # dashdot
plt.plot(x, x+7, linestyle=':') # dotted
```

- xticks, yticks: x축, y축으로 사용할 값
- xlim, ylim: x축, y축 한계

### 4-3-3. 히스토그램
- 도수분포표를 그래프로 나타낸 것
- 가로축: 계급(변수의 구간, bin)
- 세로축: 도수(빈도수, frequency)
- 전체 총량: n

##### 예시 데이터
- x1: 평균(100), 표준편차(15)인 정규분포
- x2: 평균(130), 표준편차(15)인 정규분포
- 도수(50), 확률 밀도가 아닌 빈도로 표기

```python
# 그래프 데이터
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

# 축 그리기
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)

# 그래프 그리기
patches = ax1.hist(x1, bins=50, density=False) # bins=50: x값을 총 50개 구간으로 나눔
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)

ax1.xaxis.set_ticks_position('bottom') # x축의 눈금을 아래 표시
ax1.yaxis.set_ticks_position('left') # y축의 눈금을 왼쪽에 표시

# 라벨, 타이틀 달기
plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
ax1.set_title('Two Frequency Distributions')

# 보여주기
plt.show()
```

##### 위에서의 예제 데이터를 히스토그램으로

## 6. Heatmap
- 방대한 양의 데이터와 현상을 수치에 따른 색상으로 나타내는 것
- 모두 2차원으로 시각화하여 표현
- pivot: '어떤 축, 점을 기준으로 바꾼다' 라는 뜻

```python
# pandas의 dataframe의 pivot() 메서드 사용
# flights의 탑승객 수를 year과 month로 pivot 해보자

pivot = flights.pivot(index='year', columns='month', values='passengers')
pivot
```

```python
flights.head()
```

원래 위와 같던 flights 데이터 프레임이 year과 month로 pivot됨을 알 수 있다.

```python
sns.heatmap(pivot)
```

```python
# annot=True: 값 출력
# fmt='d': 정수로 출력

sns.heatmap(pivot, linewidths=.2, annot=True, fmt='d')
```
